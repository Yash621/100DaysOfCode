```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression








```

```python
make_regression?
```

```python
X,y,coef=make_regression(n_features=1,noise=20,bias=100,random_state=42,coef=True)
coef
plt.scatter(X,y)
```

```python
y_gen =X*coef + 100
plt.scatter(X,y)
plt.plot(X,y_gen)
```

```python
from sklearn.linear_model import LinearRegression

model =LinearRegression()

model.fit(X,y)
```

```python
model.coef_ ,model.intercept_
```

```python
y_gene=model.coef_*X +model.intercept_
plt.scatter(X,y)
plt.plot(X,y_gen,label="Pre")
plt.plot(X,model.predict(X),label="LR")
plt.legend()
```
